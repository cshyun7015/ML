```python
    """
    Return true iff u is equal to v.
    Because of sparse representation, it is not enough to compare dictionaries

    Consider using brackets notation u[...] and v[...] in your procedure
    to access entries of the input vectors.  This avoids some sparsity bugs.

    >>> Vec({'a', 'b', 'c'}, {'a':0}) == Vec({'a', 'b', 'c'}, {'b':0})
    True
    >>> Vec({'a', 'b', 'c'}, {'a': 0}) == Vec({'a', 'b', 'c'}, {})
    True
    >>> Vec({'a', 'b', 'c'}, {}) == Vec({'a', 'b', 'c'}, {'a': 0})
    True

    Be sure that equal(u, v) checks equalities for all keys from u.f and v.f even if
    some keys in u.f do not exist in v.f (or vice versa)

    >>> Vec({'x','y','z'},{'y':1,'x':2}) == Vec({'x','y','z'},{'y':1,'z':0})
    False
    >>> Vec({'a','b','c'}, {'a':0,'c':1}) == Vec({'a','b','c'}, {'a':0,'c':1,'b':4})
    False
    >>> Vec({'a','b','c'}, {'a':0,'c':1,'b':4}) == Vec({'a','b','c'}, {'a':0,'c':1})
    False

    The keys matter:
    >>> Vec({'a','b'},{'a':1}) == Vec({'a','b'},{'b':1})
    False

    The values matter:
    >>> Vec({'a','b'},{'a':1}) == Vec({'a','b'},{'a':2})
    False
    """
    
print(Vec({'a', 'b', 'c'}, {'a':0}) == Vec({'a', 'b', 'c'}, {'b':0}))
print(Vec({'a', 'b', 'c'}, {'a': 0}) == Vec({'a', 'b', 'c'}, {}))
print(Vec({'a', 'b', 'c'}, {}) == Vec({'a', 'b', 'c'}, {'a': 0}))
print("")
print(Vec({'x','y','z'},{'y':1,'x':2}) == Vec({'x','y','z'},{'y':1,'z':0}))
print(Vec({'a','b','c'}, {'a':0,'c':1}) == Vec({'a','b','c'}, {'a':0,'c':1,'b':4}))
print(Vec({'a','b','c'}, {'a':0,'c':1,'b':4}) == Vec({'a','b','c'}, {'a':0,'c':1}))
print("")
print(Vec({'a','b'},{'a':1}) == Vec({'a','b'},{'b':1}))
print(Vec({'a','b'},{'a':1}) == Vec({'a','b'},{'a':2}))
print("")
print(Vec({'a','b'},{'a':1}) == Vec({'a','b'},{'a':1}))
```

```python
    """
    Returns the dot product of the two vectors.

    Consider using brackets notation u[...] and v[...] in your procedure
    to access entries of the input vectors.  This avoids some sparsity bugs.

    >>> u1 = Vec({'a','b'}, {'a':1, 'b':2})
    >>> u2 = Vec({'a','b'}, {'b':2, 'a':1})
    >>> u1*u2
    5
    >>> u1 == Vec({'a','b'}, {'a':1, 'b':2})
    True
    >>> u2 == Vec({'a','b'}, {'b':2, 'a':1})
    True
    >>> v1 = Vec({'p','q','r','s'}, {'p':2,'s':3,'q':-1,'r':0})
    >>> v2 = Vec({'p','q','r','s'}, {'p':-2,'r':5})
    >>> v1*v2
    -4
    >>> w1 = Vec({'a','b','c'}, {'a':2,'b':3,'c':4})
    >>> w2 = Vec({'a','b','c'}, {'a':12,'b':8,'c':6})
    >>> w1*w2
    72

    The pairwise products should not be collected in a set before summing
    because a set eliminates duplicates
    >>> v1 = Vec({1, 2}, {1 : 3, 2 : 6})
    >>> v2 = Vec({1, 2}, {1 : 2, 2 : 1})
    >>> v1 * v2
    12
    """
    
u1 = Vec({'a','b'}, {'a':1, 'b':2})
u2 = Vec({'a','b'}, {'b':2, 'a':1})
print(u1*u2)
print(u1 == Vec({'a','b'}, {'a':1, 'b':2}))
print(u2 == Vec({'a','b'}, {'b':2, 'a':1}))
v1 = Vec({'p','q','r','s'}, {'p':2,'s':3,'q':-1,'r':0})
v2 = Vec({'p','q','r','s'}, {'p':-2,'r':5})
print(v1*v2)
w1 = Vec({'a','b','c'}, {'a':2,'b':3,'c':4})
w2 = Vec({'a','b','c'}, {'a':12,'b':8,'c':6})
print(w1*w2)
v1 = Vec({1, 2}, {1 : 3, 2 : 6})
v2 = Vec({1, 2}, {1 : 2, 2 : 1})
print(v1 * v2)
```

```python
    """
    Returns the scalar-vector product alpha times v.

    Consider using brackets notation v[...] in your procedure
    to access entries of the input vector.  This avoids some sparsity bugs.

    >>> zero = Vec({'x','y','z','w'}, {})
    >>> u = Vec({'x','y','z','w'},{'x':1,'y':2,'z':3,'w':4})
    >>> 0*u == zero
    True
    >>> 1*u == u
    True
    >>> 0.5*u == Vec({'x','y','z','w'},{'x':0.5,'y':1,'z':1.5,'w':2})
    True
    >>> u == Vec({'x','y','z','w'},{'x':1,'y':2,'z':3,'w':4})
    True
    """
    
zero = Vec({'x','y','z','w'}, {})
u = Vec({'x','y','z','w'},{'x':1,'y':2,'z':3,'w':4})
print(0*u == zero)
print(1*u == u)
print(0.5*u == Vec({'x','y','z','w'},{'x':0.5,'y':1,'z':1.5,'w':2}))
print(u == Vec({'x','y','z','w'},{'x':1,'y':2,'z':3,'w':4}))
```

```python
# Task 3.12.1
f = open('voting_record_dump109.txt')
mylist = list(f)

def create_voting_dict(strlist):
    dic = {}
    for l in strlist:
        t = l.split()
        dic[t[0]] = [int(x) for x in t[3::]]
    return dic

create_voting_dict(mylist)
```

```python
# Task 3.12.2
def policy_compare(sen_a, sen_b, voting_dict):
    a = voting_dict[sen_a]
    b = voting_dict[sen_b]
    return sum([x*y for (x, y) in zip(a, b)])

policy_compare('Akaka', 'Alexander', create_voting_dict(mylist))
```

```python
# Task 3.12.3
def most_similar(sen, voting_dict):
    a = voting_dict[sen]
    m = -100
    s = ''
    for (k, v) in voting_dict.items():
        if sen != k:
            r = sum([x*y for (x, y) in zip(a, v)])
            #print(k, r)
            if (r > m): 
                m = r
                s = k
    return s
    
most_similar('Alexander', create_voting_dict(mylist))
```

```python
# Task 3.12.4
def least_similar(sen, voting_dict):
    a = voting_dict[sen]
    m = 100
    s = ''
    for (k, v) in voting_dict.items():
        if sen != k:
            r = sum([x*y for (x, y) in zip(a, v)])
            #print(k, r)
            if (r < m): 
                m = r
                s = k
    return s

least_similar('Alexander', create_voting_dict(mylist))
```

```python
# Task 3.12.7
def find_average_similarity(sen, sen_set, voting_dict):
    a = voting_dict[sen]
    sss = 0
    for ss in sen_set:
        b = voting_dict[ss]
        sss = sss + sum([x*y for (x, y) in zip(a, b)])
    return sss / len(sen_set)

find_average_similarity('Alexander', ['Chafee', 'Santorum'], create_voting_dict(mylist))
```

```python
ss = [l.split()[0] for l in mylist if l.split()[1] == 'D']

find_average_similarity('Alexander', ss, create_voting_dict(mylist))

m = -100
s = ''
for x in [l.split()[0] for l in mylist]:
    r = find_average_similarity(x, ss, create_voting_dict(mylist))
    if (r > m):
        m = r
        s = x
print(s)
```

```python
# Task 3.12.8
def find_average_record(sen_set, voting_dict):
    ssa = [0 for x in range(len(sen_set))]
    for ss in sen_set:
        ssa = [x+y for (x, y) in zip(ssa, voting_dict[ss])]
    return [x / len(sen_set) for x in ssa]

ss = [l.split()[0] for l in mylist if l.split()[1] == 'D']
average_Democrat_record = find_average_record(ss, create_voting_dict(mylist))

a = average_Democrat_record
m = -100
s = ''
for (k, v) in create_voting_dict(mylist).items():
    r = sum([x*y for (x, y) in zip(a, v)])
    if (r > m): 
        m = r
        s = k
print(s)
```

```python
# Task 3.12.9
def bitter_rivals(voting_dict):
    sen = [x for x in voting_dict.keys()]
    m = 100
    s1 = ''
    s2 = ''
    for x in sen:
        for y in sen:
            r = policy_compare(x, y, voting_dict)
            if (r < m):
                m = r
                s1 = x
                s2 = y
    return s1, s2

bitter_rivals(create_voting_dict(mylist))
```

```python
# An expression
# α1v1 +···+ αnvn
# is a linear combination of the vectors v1,...,vn.
# The scalars α1,...,αn are the coefficients of the linear combination
```

```python
# Definition: The set of all linear combinations of some vectors
# v1,...,vn is called the span of these vectors
# Written Span{v1,...,vn}
```

```python
# Definition: Let V be a set of vectors.  If v1,...,vn are vectors such that
# V = Span{v1,...,vn} then
#  we say {v1,...,vn} is a generating set for V;
#  we refer to the vectors v1,...,vn as generators for V
```

```python
# Definition:
# If c is a vector and V is a vector space then c+V is called an affine space
```

```python
# Definition:
# A linear combination γu1+αu2+βu3 where γ+α+β = 1 is an affine combination.
```

```python
# Definition: For vectors v1,...,vn over R, a linear combination
# α1v1 + ··· + αnvn is a convex  combination 
# if the coefficients are all nonnegative and they sum to 1.
```

```python
# Problem 4.8.1
from vec import *

def vec_select(veclist, k):
    return [v for v in veclist if v[k] == 0]

vec_select([Vec({1,2},{1:1}), Vec({1,2},{2:1})],1)
```

```python
def vec_sum(veclist, D):
    return sum(veclist)

vec_sum([Vec({1,2},{1:1,2:4}), Vec({1,2},{2:1})], {1,2})
```

```python
def vec_select_sum(D, veclist, k):
    return vec_sum(vec_select(veclist, k),D)

vec_select_sum({1,2}, [Vec({1,2},{1:1,2:4}), Vec({1,2},{2:1})], 1)
```
