```python
# Task 14.13.1
from vec import *

def main_constraint(i, a_i, d_i, features):
    f = {}
    f.update(a_i.f)
    f.update({'gamma':-1 if d_i == 1 else 1})
    f.update({i:1})
    return Vec(features, f)

main_constraint(862722, Vec({'f1','f2'},{'f1':33}), -1, b_.D.union({'f1','f2','gamma'}))
```

```python
# Task 14.13.2
from matutil import *

def make_matrix(feature_vectors, diagnoses, features):
    mc = {k:main_constraint(k,v,diagnoses[k],features) for k,v in feature_vectors.items()}
    nc = {-1*k:Vec(features, {k:1}) for k in feature_vectors.keys()}
    f = {}
    f.update(mc)
    f.update(nc)
    return rowdict2mat(f)

A = make_matrix(mat2rowdict(M_), b_, M_.D[1].union(b_.D).union({'gamma'}))
```

```python
# Task 14.13.3

def return_b():
    return Vec(A.D[0], {l:1 for l in A.D[0] if l > 0})

b = return_b()
```
