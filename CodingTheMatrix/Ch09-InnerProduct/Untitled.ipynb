```python
# Task 9.4.2

def signum(u):
    f = {}
    for k in u.D:
        if u[k] >= 0: f[k] = 1
        else: f[k] = -1
    return Vec(u.D, f)

signum(Vec({'A','B'}, {'A':3,'B':-2}))
```

```python
# Task 9.4.3

def fraction_wrong(A, b, w):
    x = A*w
    xs = signum(x)
    return sum([1 for k in b.D if b[k]*xs[k] < 0])/len(b.D)

A = Mat(({1,2,3},{'A','B','C'}),{(1,'A'):3,(2,'B'):3,(3,'C'):3})
b = Vec({1,2,3},{1:-1, 2:-1, 3:1})
w = Vec({'A','B','C'},{'A':1, 'B':1, 'C':-1})

fraction_wrong(A,b,w)
```

```python
# Task 9.4.4

def loss(A, b, w):
    x = A*w
    return (x-b)*(x-b)

A = Mat(({1,2,3},{'A','B','C'}),{(1,'A'):3,(2,'B'):3,(3,'C'):3})
b = Vec({1,2,3},{1:-1, 2:-1, 3:1})
w = Vec({'A','B','C'},{'A':1, 'B':1, 'C':-1})

loss(A,b,w)
```

```python
# Task 9.4.9

def find_grid(A, b, w):
    return 2*(A*w-b)*A

A, b = read_training_data('train.data')
w = Vec(A.D[1], {k:1 for k in A.D[1]})
find_grid(A,b,w)
```

```python
# Task 9.4.10

def gradient_descent_step(A, b, w, sigma):
    H = sigma*find_grid(A,b,w)
    return w - H

A, b = read_training_data('train.data')
w = Vec(A.D[1], {k:1 for k in A.D[1]})
gradient_descent_step(A, b, w, 0.01)
```

```python
# Task 9.4.11

def gradient_descent(A, b, w, sigma, T):
    for i in range(T):
        if i % 30 == 0:
            print('L(w), wrong rate = ', loss(A,b,w), fraction_wrong(A, b, w))
        w = gradient_descent_step(A, b, w, sigma)
    return w

A, b = read_training_data('train.data')
w = Vec(A.D[1], {k:1 for k in A.D[1]})
gradient_descent(A, b, w, 10**-9, 3000)
```
