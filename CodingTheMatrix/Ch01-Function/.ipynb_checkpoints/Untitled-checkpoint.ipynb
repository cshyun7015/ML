```python
# Task 1.6.6

def makeInverseIndex(strlist):
    L = list(enumerate(strlist))
    dic = {}
    for i in L:
        for ii in i[1].split():
            s = dic[ii] | {i[0]} if ii in dic else {i[0]}
            dic[ii] = s
    return dic
    
f = open('stories_small.txt')
stories = list(f)
makeInverseIndex(stories)
```

```python
# Task 1.6.7
def orSearch(inverseIndex, query):
    s = inverseIndex[query[0]]
    for i in query:
        s.update(inverseIndex[i])
    return s

II = makeInverseIndex(stories)
orSearch(II, ['prep', 'course'])
```

```python
# Task 1.6.8
def andSearch(inverseIndex, query):
    s = inverseIndex[query[0]]
    for i in query:
        s.intersection_update(inverseIndex[i])
    return s

II = makeInverseIndex(stories)
andSearch(II, ['prep', 'course'])
```
